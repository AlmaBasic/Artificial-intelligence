1. Napisati code u Python-u koji za unesenu vrijednost N ispise sve Fibonaccijeve brojeve od prvog do N-tog.

```python
def fibonacci(n):
    fib_sequence = [0, 1]  # Prva dva Fibonacci broja
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence[:n]

N = int(input("Unesite broj N: "))

print("Fibonacci brojevi od 1 do", N, "su:")
print(fibonacci(N))
```

2. Napisati code u Python-u koji za unesenu vrijednost N generise dvije liste od po N nasumicnih brojeva, realnog tipa. Program potom treba kreirati novu listu ciji je i-ti element jednak zbiru brojeva na i-tom mjestu prve i druge liste.

```python
import random

def generate_random_list(N):
    return [random.random() for _ in range(N)]

def sum_lists(list1, list2):
    return [x + y for x, y in zip(list1, list2)]

N = int(input("Unesite broj N: "))

lista1 = generate_random_list(N)
lista2 = generate_random_list(N)

nova_lista = sum_lists(lista1, lista2)

print("Prva lista:", lista1)
print("Druga lista:", lista2)
print("Zbir prve i druge liste:", nova_lista)
```

3. Koristeci biblioteku numpy od korisnika traziti unos 4 realna broja te ih prikazati u matrici 2x2. Potom ispisati originalnu matricu, te transponovanu matricu. Ispisati i rezultat metode flatten().

```python
import numpy as np


unos = []
for i in range(4):
    unos.append(float(input("Unesite realan broj: ")))

# Kreiranje matrice 2x2 pomoću NumPy
matrica = np.array(unos).reshape(2, 2)

print("Originalna matrica:")
print(matrica)

print("\nTransponovana matrica:")
print(matrica.T)

print("\nRezultat metode flatten():")
print(matrica.flatten())
```

6. Vrijednosti dobivene u zadatku 5 prikazati graficki koristeci biblioteku matplotlib. Na x-osi se treba nalaziti dimenzija liste (N), a na y-osi vrijeme izvrsavanja code-a. Na istom grafiku potrebnu je prikazati vrijeme izvrsavanja i koristeci built-in list i koristeci numpy. Na grafiku dodati: title, x-axis title, y-axis title i legend.

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def generate_random_list(N):
    return np.random.rand(N)

def sum_lists(list1, list2):
    return list1 + list2

# Vrijednosti N koje želimo testirati
vrijednosti_N = [10, 100, 1000, 10000, 100000, 1000000]

# Liste za pohranu vremena izvršavanja
vrijeme_list = []
vrijeme_numpy = []

# Mjerenje vremena za zadatke 2 i 4
for N in vrijednosti_N:
    start_time = time.time()
    lista1 = generate_random_list(N)
    lista2 = generate_random_list(N)
    nova_lista = sum_lists(lista1, lista2)
    end_time = time.time()
    vrijeme_list.append(end_time - start_time)

    start_time = time.time()
    unos = np.random.rand(N)
    matrica = unos.reshape(int(N/2), 2)
    transponovana_matrica = matrica.T
    rezultat_flatten = matrica.flatten()
    end_time = time.time()
    vrijeme_numpy.append(end_time - start_time)

# Grafikon
plt.plot(vrijednosti_N, vrijeme_list, label='Built-in list')
plt.plot(vrijednosti_N, vrijeme_numpy, label='NumPy')
plt.title('Vrijeme izvršavanja za različite vrijednosti N')
plt.xlabel('Veličina liste (N)')
plt.ylabel('Vrijeme izvršavanja (s)')
plt.legend()
plt.xscale('log')  # Logaritamska skala za bolju čitljivost na x-osi
plt.yscale('log')  # Logaritamska skala za bolju čitljivost na y-osi
plt.grid(True)     # Dodavanje mreže na grafikon
plt.show()
```

7. Koristeci matplotlib nacrtati grafik funkcije f(x) = e^(−x) · sin(x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicija funkcije f(x)
def f(x):
    return np.exp(-x) * np.sin(x)

# Generiranje niza x vrijednosti
x_values = np.linspace(0, 10, 100)  # od 0 do 10 sa 100 tačaka

# Izračunavanje odgovarajućih vrijednosti f(x)
y_values = f(x_values)

# Plotanje grafika
plt.plot(x_values, y_values, label='$f(x) = e^{-x} \cdot \sin(x)$')
plt.title('Grafik funkcije $f(x) = e^{-x} \cdot \sin(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
